```python
import cv2
import matplotlib.pyplot as plt
```

```python
webcam_ids = []

for i in range(10):
    cap = cv2.VideoCapture(i)
    ret, img = cap.read()
    if ret:
        webcam_ids.append(i)
    cap.release()

print(webcam_ids)
    
```
